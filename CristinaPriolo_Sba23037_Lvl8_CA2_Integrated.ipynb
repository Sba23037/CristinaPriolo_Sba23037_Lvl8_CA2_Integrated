# Encoding:

I notice that over18 don't have variance, so it is not considered for the next step that is Encoding the categorical variable into numerical with LabelENcoder 

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
cn1 = list(dataset['Attrition'].values)
le.fit(list(set(cn1)))
num_cn1 = list(le.transform(cn1))
dataset['Attrition'] = num_cn1

le = LabelEncoder()
cn1 = list(dataset['BusinessTravel'].values)
le.fit(list(set(cn1)))
num_cn1 = list(le.transform(cn1))
dataset['BusinessTravel'] = num_cn1

le = LabelEncoder()
cn1 = list(dataset['EducationField'].values)
le.fit(list(set(cn1)))
num_cn1 = list(le.transform(cn1))
dataset['EducationField'] = num_cn1

le = LabelEncoder()
cn1 = list(dataset['Department'].values)
le.fit(list(set(cn1)))
num_cn1 = list(le.transform(cn1))
dataset['Department'] = num_cn1

le = LabelEncoder()
cn1 = list(dataset['Gender'].values)
le.fit(list(set(cn1)))
num_cn1 = list(le.transform(cn1))
dataset['Gender'] = num_cn1

le = LabelEncoder()
cn3 = list(dataset['JobRole'].values)
le.fit(list(set(cn3)))
num_cn3 = list(le.transform(cn3))
dataset['JobRole'] = num_cn3

le = LabelEncoder()
cn2 = list(dataset['MaritalStatus'].values)
le.fit(list(set(cn2)))
num_cn2 = list(le.transform(cn2))
dataset['MaritalStatus'] = num_cn2


le = LabelEncoder()
cn3 = list(dataset['OverTime'].values)
le.fit(list(set(cn3)))
num_cn3 = list(le.transform(cn3))
dataset['OverTime'] = num_cn3
le = LabelEncoder()
cn3 = list(dataset['Over18'].values)
le.fit(list(set(cn3)))
num_cn3 = list(le.transform(cn3))
dataset['Over18'] = num_cn3
```

Now the categorical variable bacame binary and it needs to be started handling the missing variable inside before proceeding with the next steps.If the missing variable will be dropped the dataset will be left with no enought obeservation. To handle the missing variable no there is the risk of the choice so it is wise to use different method for each features. 

For Binary Variable nit couldn't be used in this case the median due that for example in the column gender if we fill with the median won't be accurate to identify the male or female. it will be filed with the mode the risk is if in the dataset are present 10 men, 5 women and 10 missing variable if it is filled with the mode ( th emost frequent variable) it will be assumed that the missing variable will be male and this could be not accurate. 

```python
from sklearn.impute import SimpleImputer
imputer= SimpleImputer(strategy='most_frequent')
dataset['binary_variable']= imputer.fit_transform(dataset[['Gender']])
```

```python
from sklearn.impute import SimpleImputer
imputer= SimpleImputer(strategy='most_frequent')
dataset['binary_variable']= imputer.fit_transform(dataset[['MaritalStatus']])
```

```python
dataset.isnull().sum()
```

The rest can be filled with the median: 

```python
dataset.fillna(dataset.median(),inplace=True)
```
